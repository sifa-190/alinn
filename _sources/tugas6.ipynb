# TRANSFORMASI LINIER

## Pengertian :

Transformasi adalah pemetaan T dari ruang vektor V ke ruang vektor W.

Cara menentukan tranformasi linier :
- pusat titik tidak berubah
- garis lurus ditetapkan sebagai lurus

## Definsi
Transformasi adalah pemetaan T dari ruang vektor V ke ruang vektor W .

**v ∈ V Transformation −−−−−−−−→ w = T (v) ∈ W**

Secara umum pemetaan T adalah transformasi linier jika

$$ T( u+v) = T(u) + T(v) $$
$$ T (\alpha v) = \alpha T (v) $$

- contoh:carilah matriks standar $A$ linear transformation dari $\mathbb{R}^3$ to $\mathbb{R}^2$ sehingga

$$
\begin{gathered}

## Refleksi

Refleksi atau pencerminan selain terhadap garis vertikal atau garis horizontal, juga dapat dilakukan pencerminan terhadap garis yang lainnya yaitu terhdap garis $y=m x+c$ atau terhadap garis $a x+b y+c=0$. Untuk pencerminan tipe ini, akan kita bahas pada artikel lainnya secara lebih mendalam. Silahkan teman-teman ikuti link: "Pencerminan terhadapgaris $y=m x+c$ ".

Berikut adalah matriks transformasi untuk refleksi berdasarkan garis sebagai cerminnya yaitu :
). Pencerminan terhadap sumbu X
$$
A(a, b) \longrightarrow A^{\prime}(a,-b) \text {. }
$$

Matriksnya : $M=\left(\begin{array}{cc}1 & 0 \\ 0 & -1\end{array}\right)$
enghitungan: $\binom{a^{\prime}}{b^{\prime}}=\left(\begin{array}{cc}1 & 0 \\ 0 & -1\end{array}\right)\binom{a}{b}$
). Pencerminan terhadap sumbu $Y$
$$
A(a, b) \longrightarrow A^{\prime}(-a, b) \text {. }
$$

Matriksnya : $M=\left(\begin{array}{cc}-1 & 0 \\ 0 & 1\end{array}\right)$
enghitungan : $\binom{a^{\prime}}{b^{\prime}}=\left(\begin{array}{cc}-1 & 0 \\ 0 & 1\end{array}\right)\binom{a}{b}$
). Pencerminan terhadap garis $y=x$
$$
A(a, b) \longrightarrow A^{\prime}(b, a)
$$

Matriksnya : $M=\left(\begin{array}{ll}0 & 1 \\ 1 & 0\end{array}\right)$
enghitungan: $\binom{a^{\prime}}{b^{\prime}}=\left(\begin{array}{ll}0 & 1 \\ 1 & 0\end{array}\right)\binom{a}{b}$

*). Pencerminan terhadap garis $y=-x$
$$
A(a, b) \longrightarrow A^{\prime}(-b,-a)
$$

Matriksnya : $M=\left(\begin{array}{cc}0 & -1 \\ -1 & 0\end{array}\right)$
Penghitungan : $\binom{a^{\prime}}{b^{\prime}}=\left(\begin{array}{cc}0 & -1 \\ -1 & 0\end{array}\right)\binom{a}{b}$
*). Pencerminan terhadap titik asal yaitu pusat koordinat $(0,0)$
$$
A(a, b) \longrightarrow A^{\prime}(-a,-b)
$$

Matriksnya : $M=\left(\begin{array}{cc}-1 & 0 \\ 0 & -1\end{array}\right)$
Penghitungan : $\binom{a^{\prime}}{b^{\prime}}=\left(\begin{array}{cc}-1 & 0 \\ 0 & -1\end{array}\right)\binom{a}{b}$
*). Pencerminan terhadap garis $x=h$
$$
\begin{aligned}
& A(a, b) \\
& a^{\prime}=2 h-a A^{\prime}(2 h-a, b) \\
& b^{\prime}=b \rightarrow a^{\prime}=(-1 . a+0 . b)+2 h \\
&
\end{aligned}
$$

Matriksnya : $M=\left(\begin{array}{cc}-1 & 0 \\ 0 & 1\end{array}\right)$
Penghitungan : $\binom{a^{\prime}}{b^{\prime}}=\left(\begin{array}{cc}-1 & 0 \\ 0 & 1\end{array}\right)\binom{a}{b}+\binom{2 h}{0}$
*). Pencerminan terhadap garis $y=k$

*). Pencerminan terhadap garis $y=k$
$$
\begin{aligned}
& A(a, b) \longrightarrow A^{\prime}(a, 2 k-b) . \\
& a^{\prime}=a \rightarrow a^{\prime}=(1 . a+0 . b)+0 \\
& b^{\prime}=2 k-b \rightarrow b^{\prime}=(0 . a+-1 . b)+2 k
\end{aligned}
$$

Matriksnya : $M=\left(\begin{array}{cc}1 & 0 \\ 0 & -1\end{array}\right)$
Penghitungan : $\binom{a^{\prime}}{b^{\prime}}=\left(\begin{array}{cc}1 & 0 \\ 0 & -1\end{array}\right)\binom{a}{b}+\binom{0}{2 k}$


Tranformasi Linier dengan Refleksi atau pencerminan pada tranformasi linear adalah adalah salah satu metode tranformasi linear yang digunakan untuk mencerminkan suatu Node dengan fungsi matrix. Berikut adalah sifat-sifat Refleksi:

1. Objek yang direfleksikan tidak mengalami perubahan benntuk dan ukuran
2. Jarak objek dari cermin (sumbu X/Y) adalah sama dengan jarak bayangan dengan cermin tersbut.

Contoh :

Misalkan refleksi dari titik A(1,1)dan bentuk matriksnya yaitu :

$ A  \begin{pmatrix}
1 & 2 \\
2 & 1 \\
\end{pmatrix} $
